# Feature EngineeringÂ¶
**'Year': Extracted the year from the 'Date' column.

'Month': Extracted the month from the 'Date' column.

'Day': Extracted the day of the month from the 'Date' column.

'DayOfWeek': Extracted the day of the week from the 'Date' column (0 - Monday, 6 - Sunday).

'Quarter': Extracted the quarter of the year from the 'Date' column.

'Season': Created a new column 'Season' based on the month using a custom function.**

```python
#convert date in to datetime format
df['Date'] = pd.to_datetime(df["Date"])

df = df.sort_values(by='Date')

# Extract year, month, day, day of the week, quarter, and season as new features
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['DayOfWeek'] = df['Date'].dt.dayofweek
df['Quarter'] = df['Date'].dt.quarter

# Define a function to map months to seasons
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Autumn'

# Apply the function to create a new 'Season' column
df['Season'] = df['Month'].apply(get_season)
df.head(10)
```

**Scale the data as a change from the initial data.**

```python
close_columns = ['Date', 'Crypto', 'Close']
close_df = df[close_columns].sort_values(by=['Date', 'Crypto']).reset_index(drop=True)
currencies = sorted(close_df['Crypto'].unique().tolist())

def get_scaled(currency: str, arg_df: pd.DataFrame) -> pd.DataFrame:
    result_df = arg_df[arg_df['Crypto'] == currency].copy(deep=True)
    min_date = result_df['Date'].min()
    result_df['Close'] = result_df['Close'].astype(float)
    scale_value = result_df[result_df['Date'] == min_date]['Close'].astype(float)
    result_df['Close'] = result_df['Close'].apply(func=lambda x: x/scale_value)
    return result_df

scaled_df = pd.concat([get_scaled(currency=currency, arg_df=close_df) for currency in currencies]).sort_values(by=['Date', 'Crypto']).reset_index(drop=True)

plt.figure(figsize=(10, 6))
for currency in currencies:
    crypto_df = scaled_df[scaled_df['Crypto'] == currency]
    plt.plot(crypto_df['Date'], crypto_df['Close'], label=currency)
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('Close Price of Cryptocurrencies')
plt.legend()
plt.show()
```

```python
btc_df= df[df['Crypto'] == 'BTC'][['Date', 'Close']].set_index('Date').asfreq('D')
btc_test_df = btc_df['2023-05-01':]
btc_df = btc_df[:'2023-05-01']
btc_df[btc_df['Close'].isna()]
```

```python
## btc_friday_df = btc_df[(btc_df.index.weekday == 4)]
btc_weekly_df = btc_df.resample('W').mean()


ax_btc = btc_df.plot()
btc_weekly_df.plot(ax=ax_btc)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

seasonal_decompose(x=btc_weekly_df, model='additive' , ).plot()
print() # add a bogus print to suppress double-plotting
```

```python
from itertools import product
from statsmodels.tsa.statespace.sarimax import SARIMAX
best_aic = float('inf')
maxiter = 1500
## we need lots of iterations to converge in some cases
parameters_list = list(product(range(0,4),range(10,13)))
results = []

# our data definitely has regimes; let's focus on the more recent one
# by just looking at data since 2020
endog = btc_weekly_df[btc_weekly_df.index.year > 2020]
enforce_stationarity = False
enforce_invertibility = False
freq = 'W'

for parameter in parameters_list:
    model = SARIMAX(endog=endog, order=(parameter[0], 1, parameter[1]), enforce_stationarity=enforce_stationarity,
                   enforce_invertibility=enforce_invertibility, freq=freq).fit(disp=-1, maxiter=maxiter, )
    if model.aic < best_aic:
        best_model = model
        best_aic = model.aic
        best_parameters = parameter
    results.append([parameter, model.aic])
result_df = pd.DataFrame(results, columns = ['parameters', 'aic'])

result_df
```

```python
from seaborn import heatmap
cmap = 'RdBu' #low is good
result_df['p'] = result_df['parameters'].apply(lambda x: x[0])
result_df['q'] = result_df['parameters'].apply(lambda x: x[1])
result_df['aic_int'] = result_df['aic'].apply(func=lambda x: int(x))
heatmap(annot=True, cmap=cmap, fmt='5d',
        data= result_df[['p','q','aic_int']].sort_values(by=['p','q'], ascending= True, ).pivot(index='q', columns='p', values='aic_int'),)
# if our best case is from a corner of our search space that's probably not a good sign
```
